## Cheatsheet :
### In bulk :
 - Display the content of a register : `info register reg` (example : `info register rax`)
 - Watch content of a register : `watch $rerg` (example : `watch $rax`)
 - Set breakpoint : `b` (example: `b label`)
 - Clear display : Ctrl+L
 - Add -g to nasm to make programm debuggable for gdb
 - Either invocate with `gdb programm_to_debug` or by using the `-p` param with `pid` if you wish to debug a running process.
 - Start the programm in gdb with `run`.
 - `next` aka. `n`.
 - `step` aka. `s`.
 - `backtrace` aka. `bt`.
 - `print` aka. `p`.
 - `list` aka. `l`.
 - `continue` aka. `c`.
 - `quit`
 - `info break` to display breakpoints.
 - `info watch` to display watchpoints.
 - #### Run in Emacs :
   - [Doc](https://sourceware.org/gdb/current/onlinedocs/gdb.html/Emacs.html#Emacs)
   - [Doc (from emacs, more exhaustive)](https://www.gnu.org/software/emacs/manual/html_node/emacs/Debuggers.html)
   - Run `gdb` under Emacs with `M-x gdb`.
   - `C-c C-c` for an interrupt`C-c C-z` for a stop`.
   - `C-h m` GUD Mode
   - With `C-c` prefix command :
     - `C-c C-s` step.
     - `C-c C-n` next.
     - `C-c C-i` stepi.
     - `C-c C-f` finish.
     - `C-c C-r` continue
     - `C-c <` Go up the number of frames given by numeric argument (like `up`)
     - `C-c >` Go down the number of frames given by numeric argument (like `down`)
     - `C-x Space` (`gud-break`) sets a breakpoint in source.
     - `M-x speedbar` display backtrace in separate frame.
     - type `f` in GDB buffer to get frame display.
     - 
 - #### Display :
   - `set width 70`

## The manual :
### 2 Getting in and out of `GDB`.
About how to start `gdb` and quit it.

#### 2.1 Invoking `GDB` :
##### `gdb` alone :
If you do so, GDB will expect an executable file you'll be able to specify with either `file` or `exec-file`.
##### With an executable as argument :
`gdb executable` 
###### With a "core file" :
`gdb executable core`
https://askubuntu.com/questions/1349047/where-do-i-find-core-dump-files-and-how-do-i-view-and-analyze-the-backtrace-st
##### Attach to a programm :
```
> gdb programm 1234
> gdb -p 1234
```

##### With arguments to pass to the executables :
`gdb --args gcc -O2 -c foo.c`
gcc is the executable the rest will be passed to.

GDB Can also be invoked with `--silent` (or `-q`/`--quiet`) option.
